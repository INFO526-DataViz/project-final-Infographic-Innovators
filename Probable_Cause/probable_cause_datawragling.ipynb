```python
import pandas as pd
file_path = "output.txt"
Probable_Cause=[]
# Open the file in write mode and write each value to the file
with open(file_path, "r") as file:
    Probable_Cause=(file.read().split("\n"))
flights=pd.read_csv("flight_crash_data_NTSB.csv")
flights.dropna(subset=["ProbableCause"],inplace=True)
flights["Probable_Cause"]=Probable_Cause[:-1]
flights['ProbableCause'] = flights['ProbableCause'].str.lower()
flights['Probable_Cause'] = flights['Probable_Cause'].str.lower()
```

```python
flights.Probable_Cause.loc[flights.Probable_Cause.str.contains("engine") |
                      flights.ProbableCause.str.contains("loss of engine power")|
                      flights.Probable_Cause.str.contains("power loss")|
                      flights.ProbableCause.str.contains("engine fail")]="loss of engine power"
flights.Probable_Cause.loc[flights.Probable_Cause=="pilot's inadequate compensation"]="crosswinds"
flights.Probable_Cause.loc[flights.ProbableCause.str.contains("inadequate maintenance")|
                      flights.ProbableCause.str.contains("inadequate preflight")
                      ]="inadequate flight maintenance"
flights.Probable_Cause.loc[flights.ProbableCause.str.contains("plane control")|
                      flights.Probable_Cause.str.contains("loss of control during")|
                      flights.Probable_Cause.str.contains("improper use of plane control")|
                      flights.Probable_Cause.str.contains("loss of airplane control")|
                      flights.Probable_Cause.str.contains("in-flight loss")|
                      flights.Probable_Cause.str.contains("loss of contro")|
                      flights.Probable_Cause.str.contains("loss of aircraft control")]="loss of plane control"
flights.Probable_Cause.loc[flights.ProbableCause.str.contains("fire")]="in-flight fire"
flights.Probable_Cause.loc[flights.Probable_Cause.str.contains("pilot's inadequate compensation")|
                      flights.Probable_Cause.str.contains("wind condit")]="crosswinds"
flights.Probable_Cause.loc[flights.ProbableCause.str.contains("aerodynamic stall")]="aerodynamic stall"
flights.Probable_Cause.loc[flights.Probable_Cause.str.contains("fuel")]="out of fuel"
flights.Probable_Cause.loc[flights.ProbableCause.str.contains("fatigue") |
                      flights.Probable_Cause.str.contains("failure of the") |
                      flights.Probable_Cause.str.contains("inadvertent")|
                      flights.Probable_Cause.str.contains("landing gear")|
                      flights.Probable_Cause.str.contains("carburetor")|
                      flights.Probable_Cause.str.contains("airplane's")|
                      flights.Probable_Cause.str.contains("airplaneâ€™s")|
                      flights.Probable_Cause.str.contains("in-flight separation")|
                      flights.Probable_Cause.str.contains("separation of")
                      ]="physical damage"
flights.Probable_Cause.loc[flights.Probable_Cause.str.contains("undetermined")]="undetermined"
flights.Probable_Cause.loc[flights.Probable_Cause.str.contains("pilot's failure to") | 
                      flights.Probable_Cause.str.contains("pilot's") |
                      flights.Probable_Cause.str.contains("flight instructor's failure") |
                      flights.Probable_Cause.str.contains("pilot")|
                      flights.Probable_Cause.str.contains("a factor wa")|
                      flights.Probable_Cause.str.contains("student pi")|
                      flights.Probable_Cause.str.contains("exclusive")|
                      flights.Probable_Cause.str.contains("factors contributing to")|
                      flights.Probable_Cause.str.contains("a factor in")|
                      flights.Probable_Cause.str.contains("a student")|
                      flights.Probable_Cause.str.contains("contributing to the accident")|
                      flights.Probable_Cause.str.contains("flight instructor's inadequate")|
                      flights.Probable_Cause.str.contains("flight instructor delayed remedial")|
                      flights.Probable_Cause.str.contains("improper")|
                      flights.Probable_Cause.str.contains("captain's failure to")|
                      flights.Probable_Cause.str.contains("improved")|
                      flights.Probable_Cause.str.contains("darkness,")|
                      flights.Probable_Cause.str.contains("a factor to")
                      ]="pilot's failure"
cause=["pilot's failure",
 'loss of engine power',
 'physical damage',
 'impact with trees',
 'inadequate flight maintenance',
 'aerodynamic stall',
 'loss of plane control',
 'crosswinds',
 'in-flight fire',
 'out of fuel',
 'collision with birds',
 'in-flight collision',
 'loss of directional',
 'undetermined']
flights.loc[~flights['Probable_Cause'].isin(cause), 'Probable_Cause'] = "Other"
```
